```python
import numpy as np

from numpy import genfromtxt

# Plotting:
import matplotlib.pyplot as plt

# Statistics:
import pandas as pd

# Visualisations:
import seaborn as sns

# Pandas for data analysis
import pandas as pd

#import morley dataset
data = np.genfromtxt('morley.csv', skip_header=1, delimiter=',', deletechars='"')

#first column removed
data = data[:,1:]

```

```python
df = data[:,2].reshape(5, 20).T

pand_df = pd.DataFrame(data = df, columns = ["C1", "C2", "C3", "C4", "C5"])

print(pand_df)
```

```python
pand_df.head()
```

```python
fig = plt.figure(figsize = (5, 5))
fig.suptitle('Morley csv Box Plot', fontsize=12, fontweight='bold')

ax = fig.add_subplot(111)
ax.boxplot(pand_df, patch_artist=True)

plt.axhline(y = 792.458, color = 'g', linestyle = '--', label='Speed')
ax.set_xlabel('Experiment Number')
ax.set_ylabel('Speed of Light')

plt.show()
```

# Exercise 3

```python
# Create a series of x vectors
xs = np.array([np.linspace(i, i + 2.0, 20) for i in range(10)])
# Create a corresponding series of y vectors.
ys = np.array([len(xs) + 1.0 * xs[i] - 2 * i for i in range(len(xs))])
# Add some noise.
ys = ys + np.random.normal(0.0, 1.0, ys.shape)

# Create global list of x and y values.
x = xs.flatten()
y = ys.flatten()

# Same plot but with best fit lines for each category.
for i in range(len(xs)):
    # Plot this category.
    plt.plot(xs[i], ys[i], '.')
    # Fit line to category.
    m, c = np.polyfit(xs[i], ys[i], 1)
    # Create steadily increasing x range for plot.
    xbestfit = np.linspace(xs[i].min(), xs[i].max(), 50)
    # Corresponding y values based on best fit line.
    ybestfit = m * xbestfit + c
    # Plot best fit line for category.
    plt.plot(xbestfit, ybestfit, color='lightgrey');
```
