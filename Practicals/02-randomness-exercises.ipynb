(10 * 9 * 8 * 7) / (4 * 3 * 2 * 1) divides into 210
meaning there is 210 outcomes of exactly 4 heads.
The formula is called the combination formula and it is C(n,r) = n!/(r!(n-r)1)
It is always a positive whole number as you are counting the amount of instances of a case, you cannot have a half case or a  negative number of cases

```python
from random import random
def flip_coin():
    r = random()
    if r > 0.5:
        return "Heads"
    else:
        return "Tails"
print(flip_coin())
```

In probability there is a 50% chance that the result is going to be heads, and since there is only one other outcome, the probability of that is 50% too. So running the same formula for tails will have similar outcomes to heads.

```python
import matplotlib.pyplot as plt
import math 
import numpy as np

# The number of coins we are to flip.
coins = 10

# Number of trials to run.
trials = 10000

# Run the trials.
heads = np.random.binomial(coins, 0.5, trials)

# Count the unique values in the returned list.
vals, counts = np.unique(heads, return_counts=True)

# Plot results in a histogram.
plt.bar(vals, counts);
```
